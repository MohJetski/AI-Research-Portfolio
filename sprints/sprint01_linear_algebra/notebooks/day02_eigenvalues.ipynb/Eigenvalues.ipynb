```python
import numpy as np

# 1. Matrix definieren
A = np.array([[4, 2], 
              [1, 3]])

# 2. Eigenwerte und Eigenvektoren berechnen
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenwerte: {eigenvalues}")
print(f"Eigenvektoren (Spalten): \n{eigenvectors}")

# 3. Verifizierung: A * v = lambda * v
v1 = eigenvectors[:, 0]
lambda1 = eigenvalues[0]

lhs = A @ v1
rhs = lambda1 * v1

print(f"\nVerifizierung f端r lambda1:")
print(f"A @ v1: {lhs}")
print(f"lambda1 * v1: {rhs}")
print(f"Stimmen sie 端berein? {np.allclose(lhs, rhs)}")
```
